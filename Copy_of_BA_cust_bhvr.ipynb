```python
mapping = {
    "Mon": '1',
    "Tue": '2',
    "Wed": '3',
    "Thu": '4',
    "Fri": '5',
    "Sat": '6',
    "Sun": '7',
}

df["flight_day"] = df["flight_day"].map(mapping)
```

```python
df["flight_day"].unique()
```

```python
df.head()
```

```python
df=df[['num_passengers',
       #'sales_channel',
       'trip_type', 'purchase_lead', 'length_of_stay', #'flight_hour',
       'flight_day', #'route',
       #'booking_origin', 'wants_extra_baggage', 'wants_preferred_seat','wants_in_flight_meals',
       'flight_duration', 'booking_complete']]
df.head()
```

```python
df['num_passengers'].unique()
```

```python
dummies=pd.get_dummies(df['trip_type'])
dummies
```

```python
df=pd.concat([dummies,df],axis='columns')
df.head()
```

```python
df=df.drop('trip_type',axis='columns')
df.head()
```

```python
df=df.drop('CircleTrip',axis='columns')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['purchase_lead']])
df['purchase_lead']=scaler.transform(df[['purchase_lead']])
```

```python
scaler.fit(df[['length_of_stay']])
df['length_of_stay']=scaler.transform(df[['length_of_stay']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('booking_complete',axis='columns')
y=df['booking_complete']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=40)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
rfc.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(rfc,x,y,cv=5)
```

```python
score=cross_val_score(rfc,x,y,cv=5)
score.mean()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted=rfc.predict(x_test)
y_predicted
```

```python
y_test
```

```python
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True)
```
